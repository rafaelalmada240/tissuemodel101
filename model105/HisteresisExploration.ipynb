```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

dt = 0.1
T = 150
N = int(T/0.01)
pinit = -1
f = np.pi/20

p_range = np.zeros(N)
p_range[0] = pinit
```

```python
t = np.arange(N)*0.01
plt.plot(t,np.floor(16*np.sin(f*t))/16)
```

```python
for i in range(N-1):
    lambda1 = 0.4*(np.round(16*np.sin(f*t[i]))/16)
    p_range[i+1] = p_range[i] - (p_range[i]**3-p_range[i]+lambda1)*dt
```

```python
plt.plot(np.floor(16*np.sin(f*t))/16,p_range,'.-')
plt.title("$p_w$ for periodic $\lambda$",fontsize = 18)
plt.xlabel("$\lambda$",fontsize = 18)
plt.grid()
#plt.legend(fontsize=16)
```
